```python
#特性之间相关性
data_corr = data.corr().abs()
print(data_corr)
plt.subplots(figsize=(20, 12))
sns.heatmap(data_corr,annot=True)
```

```python
sns.heatmap(data_corr, mask=data_corr < 0.5, cbar=False)
```

```python
cols=data.columns
threshold = 0.5

corr_list = []

size = data_corr.shape[0]

#Search for the highly correlated pairs
for i in range(0, size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) #store correlation and columns index

#排列相关性           
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

#输出结果
for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))
```

```python
# Scatter plot of only the highly correlated pairs
for v,i,j in s_corr_list:
    sns.pairplot(data, size=6, x_vars=cols[i],y_vars=cols[j] )
    plt.show()
```

开始训练模型

```python
#从原始数据中分离特征和标签
Y=data.MEDV.values
X=data.drop("MEDV",axis=1)
X.shape
colomns=X.columns
```

```python
#利用sklearn进行模型训练
#导入模块，划分训练集
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=33,test_size=0.2)
X.shape
```

```python
#对数据进行归一化
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
Y_train=ss.fit_transform(Y_train.reshape(-1,1))
Y_test=ss.transform(Y_test.reshape(-1,1))
```

使用缺省的线性回归

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
Y_test_pre=lr.predict(X_test)
Y_train_pre=lr.predict(X_train)
fs=pd.DataFrame({"columns":list(colomns),"coef":list((lr.coef_.T))})
fs.sort_values(by=["coef"],ascending=False)
```

```python
#评价模型
from sklearn.metrics import r2_score
print('train:',r2_score(Y_train,Y_train_pre))
print('test:',r2_score(Y_test,Y_test_pre))
```

```python
#绘制残差分布
fig=plt.subplots(figsize=(7,5))
plt.hist(Y_train-Y_train_pre,bins=30,label="Residuals Linear")
plt.show()
```

```python
#绘制散点图
plt.figure(figsize=(4,3))
plt.scatter(Y_train,Y_train_pre)
plt.plot([-3, 3], [-3, 3], '--k')
plt.xlabel('True price')
plt.ylabel('Predicted price')
```

```python
from sklearn.linear_model import SGDRegressor
sgdr=SGDRegressor()
sgdr.fit(X_train,Y_train)
Y_pre=sgdr.predict(X_test)
sgdr.coef_
```

```python
print('train:',sgdr.score(X_train,Y_train))
print('test:',sgdr.score(X_test,Y_test))
```

```python
from sklearn.linear_model import RidgeCV
alphas=[0.01,0.1,1,10,100]
ridge=RidgeCV(alphas=alphas,store_cv_values=True)
ridge.fit(X_train,Y_train)
Y_train_pre=ridge.predict(X_train)
Y_test_pre=ridge.predict(X_test)
print('train:',r2_score(Y_train,Y_train_pre))
print('test:',r2_score(Y_test,Y_test_pre))
```

```python
mse_mean=np.mean(ridge.cv_values_,axis=0)
plt.plot(np.log10(alphas),mse_mean.reshape(len(alphas),1))
plt.xlabel('log(alpha)')
plt.ylabel('mse')
plt.show()
print('alpha:',ridge.alpha_)
fs=pd.DataFrame({'colomns':list(colomns),"coef":list((lr.coef_.T)),'coef_ridge':list((ridge.coef_.T))})
fs.sort_values(by=['coef'],ascending=False)
```

正则化的线性回归

```python
from sklearn.linear_model import LassoCV
alphas=[0.01,0.1,1,10,100]
lasso=LassoCV(alphas=alphas)
lasso.fit(X_train,Y_train)
Y_train_pre=lasso.predict(X_train)
Y_test_pre=lasso.predict(X_test)
print('train:',r2_score(Y_train,Y_train_pre))
print('test:',r2_score(Y_test,Y_test_pre))
```

```python
mses = np.mean(lasso.mse_path_, axis = 1)
plt.plot(np.log10(lasso.alphas_), mses) 
#plt.plot(np.log10(lasso.alphas_)*np.ones(3), [0.3, 0.4, 1.0])
plt.xlabel('log(alpha)')
plt.ylabel('mse')
plt.show()    
            
print ('alpha is:', lasso.alpha_)

# 看看各特征的权重系数，系数的绝对值大小可视为该特征的重要性
fs = pd.DataFrame({"columns":list(colomns), "coef_lr":list((lr.coef_.T)), "coef_ridge":list((ridge.coef_.T)), "coef_lasso":list((lasso.coef_.T))})
fs.sort_values(by=['coef_lr'],ascending=False)
```

```python
mses = np.mean(lasso.mse_path_, axis = 1)
plt.plot(np.log10(lasso.alphas_), mses) 
#plt.plot(np.log10(lasso.alphas_)*np.ones(3), [0.3, 0.4, 1.0])
plt.xlabel('log(alpha)')
plt.ylabel('mse')
plt.show()    
            
print ('alpha is:', lasso.alpha_)
```
